## concept
Berikut adalah konsep jaringan saraf tiruan dan contoh arsitekturnya

Konsep dari jaringan saraf tiruan, seperti namanya, adalah tiruan dari jaringan saraf yang dimiliki otak yang disimulasikan oleh komputer untuk melakukan sebuah tugas seperti mencari pola. Jaringan saraf tiruan terdiri dari jaringan neuron yang terdiri dari lapisan input, satu atau lebih lapisan tersembunyi dan lapisan keluaran.
