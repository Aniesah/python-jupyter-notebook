# root
Finding a root of an equation using Python with Jupyter Notebook

## concept
Berikut adalah konsep akar dan salah satu metode cara mencarinya

### akar 
Jika diberikan sebuah persamaan polinomial.

\begin{equation}\tag{1}
f(x) = ax^n + bx^{n-1} + cx^{n-2} + ...
\end{equation}

Maka akar dari persamaan polinomial tersebut adalah nilai x yang memenuhi
\begin{equation}\tag{2}
f(x) = 0
\end{equation}

Atau dalam kata lain, x adalah solusi dari f(x) = 0

### mencari akar

Mencari akar dari sebuah polinomial dapat dilakukan dengan berbagai cara. Sebagai contoh adalah metode grafik, metode bisection dan metode secant.

### metode grafik

Pada dasarnya metode grafik adalah metode mencari akar dengan bantuan visual komputer dimana komputer akan menggambarkan grafik dari polinomial lalu dengan menggunakan grafik tersebut dapat dicari nilai akar dari polinomial tersebut yang relatif terhadap grafik polinomial.
Sebagai contoh diberikan sebuah polinomial

\begin{equation}\tag{3}
y = x^2sin(\pi x) - 2cos(2 \pi x)
\end{equation}

Menggunakan komputer dapat dicari grafiknya kemudian dengan cara manual dapat dicari titik yang dekat dengan akarnya sehingga diperoleh akar yang relatif dekat dengan akar aslinya.

### metode scan range

Metode scan range mencari akar dengan cara mencari dua nilai pada fungsi yang memberikan nilai \begin{equation}\
f(x)f(x+1) < 0
\end{equation}
kemudian jangkauan dari antara x dan x + dx menjadi jangkauan akar yang dicari
menggunakan persamaan polinomial 3 dapat dicari jangkauan akarnya sesuai dengan ketelitian yang diinginkan pada bagian kode metode scan range dibawah

### metode secant

Metode secant menggunakan hubungan iteratif dari akar 

## code
Penerapan konsep pencarian akar dengan metode yang dimaksud sebelumnya dan hasilnya adalah sebagai berikut

#### metode grafik
Kode untuk metode grafik adalah sebagai berikut

```python
import math

# define a function of x
def f(x):
    y = (x**2)*math.sin( math.pi * x) - 2 * math.cos(2 * math.pi * x)
    return y

# create data
xbeg = 0
xend = 10
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
    import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# zoom
plt.xlim([0, 5])
plt.ylim([-5, 5])

# draw a cricle
xroot = 0.78
point = Ellipse(
    xy=(xroot, 0), width=0.1, height=0.3, 
    edgecolor='r', fc='None', lw=1
)
plt.gca().add_patch(point)

# show plot result
plt.show()

print("xroot = ", xroot, sep='')
```

Secara manual akar dicari dengan memasukkan penanda pada grafik sehingga diperoleh nilai yang dekat dengan akar aslinya. Dari kode diperoleh salah satu titik yang cukup mendekati akar aslinya adalah pada x = 0.78

#### metode scan range

Kode untuk metode scan range adalah sebagai berikut.

```python
import math

# define a function of x
def f(x):
    y = (x**2)*math.sin( math.pi * x) - 2 * math.cos(2 * math.pi * x)
    return y

# create data
xbeg = 0
xend = 10
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
    import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

#scan range
dx = 0.1

root_range = []

x = xbeg
while x <= xend:
    f1 = f(x)
    f2 = f(x + dx)
    
    if f1 * f2 < 0 and x + dx <= xend:
        a_range = [x, x + dx]
        root_range.append(a_range)
        
    x += dx

print("Root range")
print(root_range)

# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# zoom
plt.xlim([0, 5])
plt.ylim([-5, 5])

# draw root range
for r in root_range:
    x1 = r[0]
    x2 = r[1]
    line = plt.Line2D((x1, x2), (0, 0), color='#f00', alpha=0.4, lw=4)
    plt.gca().add_line(line)

#show plot
plt.show()
```

Menggunakan metode scan range dapat diperoleh jangkauan yang memiliki akar dari persamaan sehingga nilai akar dari polinomial dapat diperkirakan.

#### metode secant

Kode untuk metode secant adalah sebagai berikut.

```python
import math

# define a function of x
def f(x):
    y = (x**2)*math.sin( math.pi * x) - 2 * math.cos(2 * math.pi * x)
    return y

# create data
xbeg = 0
xend = 10
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
    import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# zoom
plt.xlim([0, 5])
plt.ylim([-5, 5])

#show plot
plt.show()

# set range for secant method
n = 1
x1 = root_range[n][0]
x2 = root_range[n][1]
print("xbeg = ", x1)
print("xend = ", x2)

xrange1 = x1
xrange2 = x2

# search root
eps = 1E-3
err = 1

root = ''

while abs(err) > eps:
    x3 = x2 - f(x2) * (x2 - x1) / (f(x2) - f(x1))
    
    err = abs(f(x3))
    
    x1 = x2
    x2 = x3
    
    root = x3

print("root = ", root)
```

Menggunakan metode secant dapat diperoleh nilai yang lebih pasti dibandingkan kedua metode sebelumnya. Untuk contoh ini diperoleh bahwa akar yang berada pada jangkauan 0.7 hingga 0.7999 adalah 0.7810

Jika dibandingkan ketiganya untuk metode grafik diperoleh nilai akar x = 0.78, untuk metode scan range jangkauan akar x berada pada 0.7 < x < 0.7999 dan untuk metode secant diperoleh nilai akar x = 0.781. Sehingga ketiga metode merupakan metode yang cukup untuk mendapatkan nilai akar dari suatu persamaan polinomial sesuai dengan tingkat ketelitian masing-masing.
